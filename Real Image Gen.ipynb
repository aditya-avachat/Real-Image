# Diffusion Pipelne x Runwayml

```python
from diffusers import DiffusionPipeline
```

```python
pipe_runway = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", use_safetensors=True)
pipe_runway.to("cuda")
```

```python
image_runway = pipe_runway("a guy wearing black jeans, a white shirt, at a park").images[0]
image_runway
```

# Stable Diffusion x Runwayml

```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
pipe_runway_stable = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe_runway_stable = pipe_runway_stable.to("cuda")
```

```python
prompt = "a guy wearing black jeans, a white shirt, at a park"
image_stable_runway = pipe_runway_stable(prompt).images[0]
image_stable_runway
```

# Stable Diffusion x Realistic Vision V6 no VAE
## Best Base Model

```python
pipe_realistic = StableDiffusionPipeline.from_pretrained("SG161222/Realistic_Vision_V6.0_B1_noVAE", torch_dtype=torch.float16)
pipe_realistic = pipe_realistic.to("cuda")
```

```python
prompt = "a guy wearing black jeans, a white shirt, at a park"
image_realistic = pipe_realistic(prompt).images[0]
image_realistic
```

```python
# lets create an image according to the requirements
prompt = "a guy wearing black jeans, a white shirt, at a park"
image_realistic = pipe_realistic(prompt, height = 2048, width = 2048).images[0]
image_realistic
```
